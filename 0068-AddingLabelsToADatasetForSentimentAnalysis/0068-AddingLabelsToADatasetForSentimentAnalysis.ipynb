```python
'''
OTHER LEXICONS (BUT VADER) INCLUDE:
AFINN: AFINN is a list of English words rated for valence with an integer between -5 and +5. It is not included in the nltk library, but can be downloaded separately.
Pattern: Pattern is a web mining module that provides tools for NLP and machine learning. It includes a sentiment analysis tool that can be used for English, French, German, and Spanish texts.
TextBlob: TextBlob is a Python library for processing textual data. It includes a sentiment analysis tool that uses a pattern-based approach.
SentiWordNet: SentiWordNet is a lexical resource for opinion mining that assigns to each synset of WordNet three sentiment scores: positivity, negativity, and objectivity.
'''

import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer
nltk.download("vader_lexicon")
import pandas as pd

data = pd.read_csv(r"https://raw.githubusercontent.com/amankharwal/Website-data/master/reviews%20data.csv")
data = data.dropna()
print(data.head())
```

```python
sentiments = SentimentIntensityAnalyzer()

data["Positive"] = [sentiments.polarity_scores(i)["pos"] for i in data["Review"]]
data["Negative"] = [sentiments.polarity_scores(i)["neg"] for i in data["Review"]]
data["Neutral"] = [sentiments.polarity_scores(i)["neu"] for i in data["Review"]]
data["Compound"] = [sentiments.polarity_scores(i)["compound"] for i in data["Review"]]
data.head()
```

```python
score = data["Compound"].values
sentiment = []

for i in score:
    if i >= 0.05:
        sentiment.append("Positive")
    elif i <= -0.05:
        sentiment.append("Negative")
    else:
        sentiment.append("Neutral")

data["Sentiment"] = sentiment
data.head(50)
```
